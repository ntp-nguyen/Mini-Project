```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
x = np.zeros(10000)
y = np.zeros(10000)
for i in range(1,10000):
  k = np.random.randint(1,4)
  x[i] = 0.5*x[i-1] + k-1
  y[i] = 0.5*y[i-1]
  if k == 2:
    y[i] = y[i] + 2
plt.plot(x,y,'k.', markersize = 1)
plt.savefig('triangle.pdf')
plt.show()
```
